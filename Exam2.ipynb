```python
from numpy import *
import matplotlib.pyplot as plt # this adds matlab like plotting
%matplotlib inline 

def workFlow(Ts):
    fig, axs = plt.subplots(3)
    fig.subplots_adjust(left=0.125,right=2,bottom=0.5,top=2,hspace=0.75, wspace=1)

    A = 1.0
    OmegaA = 3*pi/2
    t = arange(0.0,4,0.01)
    B = 0.75
    OmegaB = 5*pi/2
    C = 1.25
    OmegaC = pi/2
    n = arange(0.0,4,Ts)
    x_n=A*cos(OmegaA*n)+B*cos(OmegaB*n)+C*cos(OmegaC*n)
    x_t=A*cos(OmegaA*t)+B*cos(OmegaB*t)+C*cos(OmegaC*t)
    

    axs[0].plot(t/2,x_t)
    axs[0].plot(n/2,x_n,marker='o',label='a')
    axs[0].minorticks_on()
    axs[0].grid(b=True, which='major', linewidth=2)
    axs[0].grid(True, which='both')
    axs[0].set_ylim((-3.8, 3.8))
    axs[0].set_ylabel('normalized magnitude')
    axs[0].set_xlabel('radians (multiply by pi) or multiply by 2 for seconds')
    axs[0].set_title('Original and Sampled Data')

    c = fft.fft(x_n)
    W = fft.fft(eye(size(x_n)))
    kk = range(size(c))

    axs[1].stem(kk,c,)
    axs[1].minorticks_on()
    axs[1].grid(b=True, which='major', linewidth=2)
    axs[1].grid(True, which='both')
    axs[1].set_ylabel('normalized magnitude')
    axs[1].set_xlabel('winding frequencies k')
    axs[1].set_title('DFT')

    omega = pi/(2*size(c))
    reconstructed = zeros(size(n))
    for k in kk:
        reconstructed = reconstructed + c[k]*W[k]

    reconstructed = reconstructed/size(n)

    axs[2].plot(t/2,x_t)
    axs[2].plot(n/2,reconstructed,marker='o',label='a')
    axs[2].minorticks_on()
    axs[2].grid(b=True, which='major', linewidth=2)
    axs[2].grid(True, which='both')
    axs[2].set_ylim((-3.8, 3.8))
    axs[2].set_ylabel('normalized magnitude')
    axs[2].set_xlabel('radians (multiply by pi)')
    axs[2].set_title('Plot from Winding Frequency Coefficients')

    plt.show()
```

```python
from ipywidgets import *
_ = interact(workFlow, Ts=(0,0.8,0.01))
```
